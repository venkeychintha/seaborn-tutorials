```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x='tip',y='total_bill',data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df, hue='smoker')
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

```python
##'''Categorical Plots
##=> Seaborn also helps us in doing the analysis on Categorical Data points.In this section we will discuss about
##boxplot
##violinplot
##countplot
##barplot'''
```

```python
#countplot

sns.countplot('sex', data=df)
```

```python
sns.countplot('day', data=df)
```

```python
#count plot

sns.countplot(y='sex',data=df)
```

```python
## Bar plot
sns.barplot(y='total_bill', x='smoker', data=df)
```

```python
sns.barplot(x='sex', y='total_bill', data=df)
```

```python
df.head()
```

```python
# Bax Plot

A box and whisker plot(sometimes called a boxplot)is a graph that present information from a five number summary 
```

```python
sns.boxplot('sex', 'total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient='v') #It will work on categorical data also
```

```python
#Categorized my data based on some other categories

sns.boxplot(x='total_bill', y='day', hue='smoker', data=df)
```

```python
## Violin Plot
## Violin plot Helps us to see both the distribution of data in terms of Kernel density estimation and the boxplot
```

```python
sns.violinplot(x="total_bill", y="day", data=df, palette='rainbow')
```
